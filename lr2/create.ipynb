```python
import cv2
from random import choice, randint
import numpy as np
from math import sqrt, sin, cos, radians

colors = [(255, 0, 0), (0, 128, 0), (255, 255, 0), (255, 0, 255), (0, 0, 255)]

def draw_figure(img, figure, rad, center, color):
    coords = []
    if figure == 'circle':
        cv2.circle(img, center, rad, color, thickness = -5)
        return
    elif figure == 'square':
        angle = randint(0, 90)
        for i in range(0, 4):
            x = center[0] + rad * cos(radians(angle + 90 * i))
            y = center[1] + rad * sin(radians(angle + 90 * i))
            coords.append((x, y))
    elif figure == 'triangle':
        for i in range(3):
            angle = randint(0, 90) + (i * 120)
            x = center[0] + rad * cos(radians(angle))
            y = center[1] + rad * sin(radians(angle))
            coords.append((x, y))
    elif figure == 'rectangle':
        angle = randint(30, 60)
        x1 = center[0] + rad * cos(radians(angle))
        y1 = center[1] + rad * sin(radians(angle))
        x2 = center[0] + rad * cos(radians(180 - angle))
        y2 = center[1] + rad * sin(radians(180 - angle))
        x3 = center[0] + rad * cos(radians(180 + angle))
        y3 = center[1] + rad * sin(radians(180 + angle))
        x4 = center[0] + rad * cos(radians(360 - angle))
        y4 = center[1] + rad * sin(radians(360 - angle))
        coords.append((x1, y1))
        coords.append((x2, y2))
        coords.append((x3, y3))
        coords.append((x4, y4))
    elif figure == 'pentagon':
        angle = randint(0, 90)
        for i in range(0, 5):
            x = center[0] + rad * cos(radians(angle + 72 * i))
            y = center[1] + rad * sin(radians(angle + 72 * i))
            coords.append((x, y))
    cv2.fillPoly(img, [np.asarray(coords, dtype=np.int32)], color)
    
def intersection(in_area, rad, center):
    if not in_area:
        return False
    for i in in_area:
        space = sqrt((i[0][0] - center[0]) ** 2 + (i[0][1] - center[1]) ** 2)
        if space <= (i[1] + rad):
            return True
    return False

def generate_img(number, n):
    area = []
    figure = ['circle', 'triangle', 'square', 'rectangle', 'pentagon']
    width = 600
    height = 600 
    for a in range(0, number - len(figure)):
        a = choice(figures)
        figure.append(a)
    img = np.zeros((width, height, 3), np.uint8) 
    for i in figure:
        color = choice(colors)
        rad = randint(20, 60)
        center = (randint(rad, width - rad), randint(rad,  height - rad))
        while intersection(area, rad, center):
            rad = randint(20, 60)
            center = (randint(rad, width - rad), randint(rad,  height - rad))
        area.append((center, rad))
        draw_figure(img, i, rad, center, color)
    cv2.imwrite(f"img_{n}.png", img)


if __name__ == '__main__':
    n = int(input('Количество картинок: '))
    number = int(input('Количество фигур: '))
    while n:
        generate_img(number, n)
        n -= 1
```
